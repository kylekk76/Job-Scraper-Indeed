Now we gonna use regex to split in groups the salary and extract the value to can use the numbers in a future.

```python
df[["Starter_Salary","Top_salary"]]=df["salary"].str.extract(r"(£\d{2},*\d*)?(?:.*)?(£\d*,*\d*)")
df
```

```python
df["Starter_Salary"].fillna(df["Top_salary"],inplace=True) 
df.drop("salary",axis=1)
```
